```python
# import necessary packages
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Create the text using cv2.putText

```python
def load_image():
    back_img =np.zeros((800,800))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(back_img,text='NITHIYA',org=(50,300), fontFace=font,fontScale= 5,color=(255,255,255),thickness=25,lineType=cv2.LINE_AA)
    return back_img
```

### Create the structuring element

```python
def display_image(image):
    fig = plt.figure(figsize=(13,13))
    ax = fig.add_subplot(111)
    ax.imshow(image,cmap='gray')
    plt.show()
```

```python
image=load_image()
```

```python
display_image(image)
```

### Use the Opening Operation

```python
kernel=np.ones((5,5))
```

```python
white_noise=np.random.randint(low=0,high=2,size=(800,800))
```

```python
white_noise=white_noise*255
```

```python
white_noise_image=white_noise+image
```

```python
display_image(white_noise_image)
```

```python
opening = cv2.morphologyEx(white_noise_image, cv2.MORPH_OPEN, kernel)
```

```python
display_image(opening)
```

### Use the Closing Operation

```python
black_noise=np.random.randint(low=0,high=2,size=(800,800))
```

```python
black_noise=black_noise*-255
```

```python
black_noise_image=image+black_noise
```

```python
black_noise_image[black_noise_image==-255]=0
```

```python
display_image(black_noise_image)
```

```python
closing = cv2.morphologyEx(black_noise_image, cv2.MORPH_CLOSE, kernel)
```

```python
display_image(closing)
```
